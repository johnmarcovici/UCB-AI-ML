```python
car_data_file_url = 'https://storage.googleapis.com/mledu-datasets/cars_data.csv'

feature_names = ['symboling', 'normalized-losses', 'make', 'fuel-type',
        'aspiration', 'num-doors', 'body-style', 'drive-wheels',
        'engine-location', 'wheel-base', 'length', 'width', 'height', 'weight',
        'engine-type', 'num-cylinders', 'engine-size', 'fuel-system', 'bore',
        'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
        'highway-mpg', 'price']

car_data = pd.read_csv(car_data_file_url, sep=',', names=feature_names, header=None, encoding='latin-1')
print("Output of describe = ")
car_data.describe(include = 'all')
```

```python
num_nan = car_data.price.isnull().sum()
print(num_nan)
```

```python
print("\nOutput of head = ")
car_data.head(15)
```

```python
# Filter on price = NaN when converted
car_data["price"] = pd.to_numeric(car_data["price"], errors='coerce')
car_data = car_data[car_data['price'].notna()]
```

```python
print("Output of describe after filtering = ")
car_data.describe(include = 'all')
```
