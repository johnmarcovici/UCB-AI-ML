### Problem 6: Churn by International Plan

**1 Point**

Use the `groupby` method to split the data based on customers who did and did not have international plans.  Find the average churn for each group and assign your response as a DataFrame to `intl_churn` below.  

````python
num_cont_churn = churn_df.query("intl_plan == 'no' and churn == True").churn.count()
num_cont = churn_df.query("intl_plan == 'no'").churn.count()
display([num_cont_churn, num_cont, num_cont_churn/num_cont])

num_intl_churn = churn_df.query("intl_plan == 'yes' and churn == True").churn.count()
num_intl = churn_df.query("intl_plan == 'yes'").churn.count()
display([num_intl_churn, num_intl, num_intl_churn/num_intl])

intl_churn = churn_df.groupby("intl_plan")[["churn"]].mean()
print(type(intl_churn))
display(intl_churn)
````
