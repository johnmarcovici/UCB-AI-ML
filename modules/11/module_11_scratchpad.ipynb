## Big Question



Machine learning very good at developing models and predictions when the thing being considered is like the data we have collected and modeled
- But if there is no data and we are making a prediction by guessing past can be predictive of future, ML probably won't work very well
- So ML is just one tool and there are others and need to know when is ML poorly suited for the question at hand

One alternative is **causal inference**, such as
    - Randomized control trials
    - A/B testing

Example opportunity for machine learning
- In the US, a lot of healthcare is consumed by a small number of very sick people
- Not just sick, but also have many other life challenges (income, homeless), etc..

So a possibility exists for policy makers to recognize this disparity in delivery / cost of healthcare, and get it to them earlier such that total cost is reduced

This model is called hot spotting
- Efforts to identify high utilizers of healthcare and get them help earlier

How might you gather data and approach this problem?

Not enough though to use the data we have to make predictions
- Because we need to know what would happen *if* someone were enrolled in program
- Can't just look at a prediction of high cost and see if it is reduced

In one experiment a set of patients with high spending were enrolled in a trial
- Then randomly sorted into buckets of care as usual
- Or the "high touch" interventions asserted to be beneficial at reducing cost

Experiment did show that costs for patients who got the high touch care went down dramatically
- Within 6 months of being in trial, those patients were only admitted to hospital half the time
- ... but ...
- Message was garbled, but I think it was that even patients with no interventions also saw costs go down
- So what is causal then?
- What the study showed is that when you identify a risk at a point in time, people "naturally revert back"
    - ... revert back to what?
    - they got better on their own, or treatment was effective

The data presents an interesting machine learning problem though
Want to predict not just illness and high cost, 
- but what patients can benefit from intervention
- Need to identify patients that have consistently high cost
- I think what he is saying is that *sporadic* high cost is not relevant, but *persistently* high cost

So big picture point is that models need to be tied to real world outcomes
