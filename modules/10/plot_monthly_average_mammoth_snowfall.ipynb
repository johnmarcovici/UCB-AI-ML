```python
import pandas as pd
import os
import plotly.express as px
from IPython.display import Image
import matplotlib.pyplot as plt

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(
    "./data/mammoth_mountain_snowfall_melted_table_with_monthly_average.csv"
)
df["Date"] = pd.to_datetime(df["Date"])
df = df.set_index("Date")
df
```

```python
df_monthly_average = df[["Month", "MonthlyAverage"]][:12]
df_monthly_average
```

```python
fig = px.bar(
        df_monthly_average,
        x="Month",
        y="MonthlyAverage",
        labels={"MonthlyAverage": "Average Snowfall (inches)"},
        color="MonthlyAverage",
        title="Average Snowfall Per Month, at Mammoth Mountain, CA",
        height=600,
        text_auto="d",
    )

Image(fig.update_layout(title_x=0.5).to_image(format="png", width=1400, height=500, scale=3))
```
