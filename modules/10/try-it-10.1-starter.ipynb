# Sunspots

```python
sunspots = sm.datasets.sunspots.load_pandas().data
sunspots = sunspots.set_index("YEAR")
sunspots
```

```python
df = sunspots
df.sort_index(inplace=True)
y_hist = df.loc["1900":"2000", "SUNACTIVITY"]
```

```python
y_hist.iloc[0]
```

```python
y_hist[0]
```

```python
period_guess = 12
lows = [0]
next_low_guess = period_guess
search_width = 5

while next_low_guess <= len(y_hist):   
    low_best = next_low_guess
    min_val = np.inf

    for kw in range(-search_width, search_width + 1):
        kidx = next_low_guess + kw
        if kidx < len(y_hist):
            this_val = y_hist.iloc[kidx]
            if this_val < min_val:
                min_val = this_val
                low_best = kidx

    lows.append(low_best)
    next_low_guess = low_best + period_guess
    print("Added low at index %03d with value %5.2f, next guess = %03d" % (low_best, min_val, next_low_guess))
```

```python
plt.figure(figsize=(16, 4))
plt.plot(y_hist)
for low in lows:
    if low < len(y_hist):
        plt.axvline(x=y_hist.index[low], color="red")
# plt.legend()
plt.show()
```

## Trend

```python
from statsmodels.tsa.filters.filtertools import convolution_filter
from statsmodels.tsa.seasonal import _extrapolate_trend
```

```python
avg_period = int(np.round(np.mean(np.diff(np.array(lows)))))
filt = np.ones(avg_period + 1)
filt[0] = 0.5
filt[-1] = 0.5
filt /= avg_period
filt
```

```python
trend = convolution_filter(y_hist, filt)
```

```python
k0 = 0
while np.isnan(trend.iloc[k0]):
    k0 += 1

k0_keep = k0
k0 = 0

while np.isnan(trend.iloc[k0]):
    trend.iloc[k0] = trend.iloc[k0_keep]
    k0 += 1

k1 = len(trend) - 1
while np.isnan(trend.iloc[k1]):
    k1 -= 1

k1_keep = k1
k1 = len(trend) - 1
while np.isnan(trend.iloc[k1]):
    trend.iloc[k1] = trend.iloc[k1_keep]
    k1 -= 1

display([k0_keep, k1_keep])
```

```python
plt.figure(figsize=(16, 4))
plt.plot(y_hist, "k")
plt.plot(trend, "r", linewidth=3, label="trend")
plt.legend()
plt.show()
```
