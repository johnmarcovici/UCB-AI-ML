# 14.3 Building Decision Trees in sklearn

## Fit a Tree

```python
feature_names = ["petal_length", "petal_width"]
X = df[feature_names]
y = df["species"]
decision_tree = tree.DecisionTreeClassifier(criterion="entropy").fit(X, y)
decision_tree
```

```python
four_rows = df.iloc[[18, 140, 104, 11], :]
four_rows
```

```python
decision_tree.predict(four_rows[feature_names])
```

## Plot the Decision Tree

```python
_, ax = plt.subplots(figsize=(14, 10))
obj = tree.plot_tree(
    decision_tree,
    feature_names=feature_names,
    class_names=y.unique(),
    rounded=True,
    filled=True,
    ax=ax,
)
```

### Understanding the tree labeling

- When a box is a yes/no question, the first line is the question
    - Example, first box shows: is petal length <= 2.45?
- Each box only ever uses a single feature
- samples means how many data samples input to that node
- value means samples per class
- last line class means most probable / most occurring class member
- most likely class per node is also represented with color
- darkness of color is confidence of that
- thus white means no majority class -> no color

# 14.4 Measuring Accuracy

Check the accuracy of the tree

```python
accuracy_score(y, decision_tree.predict(X)) * 100.0
```

Why is it not 100% accurate? Zoom in on one node with an "impure" state (impure = terminal but with values in more than one class)

Why couldn't that node be broken down further? Inspect the data this node received
- Notice that for the 2 features provided to the tree (petal length and width), the data are identical
- So no rule can split these further --> need more features

```python
df.query("petal_length > 2.45 and petal_width > 1.75 and petal_length < 4.85")
```

>Decision trees have perfect accuracy on training data, **except** when there are samples from different classes with the exact same features

The tendency for these trees to have nearly perfect accuracy raises a concern about overfitting

Surprisingly, the resulting tree is not vastly more complicated
- On the left, using just petal length & width
- On the right, using petal length & width and sepal length & width

**Big takeaway** - decision trees do not necessarily get more complicated or more overfit when using more features
- Especially when some of those features are good at breaking up clusters of otherwise identical samples

# 14.6 Intuitive Look at Decision Tree Boundaries

How do decision trees get made? They follow this process until every node is pure or unsplittable
- Pick the best feature x and the best split value beta
- Split the node into 2 nodes
    - One where x < beta
    - One where x >= beta

Pure node
- Has samples from only one class

Unsplittable node
- Has overlapping data points from different classes
- Therefore cannot be split

What is the "best" feature to select?
- Many are possible
- In the figure below, an arbitrary choice of petal width >= 0.8 was chosen
- It looks good, because it perfectly captures all of the setosa flowers
- But is it the best?
- To answer that, we need a definition of best from which we can drive an algorithm to choose best

# 14.7 Entropy

Define pC as the proportion per class C in a node
- In the root node below there are 34 in class 0 from a total of 110 thus p0 = 34/110 = 0.31

Now we can define entropy S: negative of sum of pc*log2(pc) for all c in C

Pop quiz! Given the node below, compute entropy

```python
values = [31, 4, 1]
pC = np.array(values) / np.sum(values)
S = -np.sum(pC * np.log2(pC))
display([pC, S])
```

But what is entropy?
- Think of it as how unpredictable a node is
- Low entropy = more predictable
- High entropy = less predictable

Notice how entropy decreases from 1.58 to 0.68 from node 1 to node 2, because setosa was very clearly defined in the second node

When a node as all samples in a single class, the entropy becomes zero

# 14.8 Using Entropy for Decision Trees

Concept: weight entropy. For a node, weighted entropy is entropy x fraction of samples in that node
- Denoted *w_S*
- Note in the root node, weighted entropy = entropy, since it contains all samples

Example, in the figure below, there are 3 samples of a total of 150 samples in a node with entropy 0.92
- So weighted entropy is 3/150*S = 0.018

Observation: as you go down the depth of the tree, entropy always goes down, as you become increasingly certain about the contents of the node
- In the figure below, delta ws = change in ws = 0.91

In this example, sum the ws in the following 2 nodes then subtract from the predecessor to get delta ws = 0.46

So what sklearn does is find the rule that maximizes delta ws

Returning to prior example of choosing which rule to separate iris species, it considered the following, and chose the last one - petal width >= 0.8, because that maximized delta ws
